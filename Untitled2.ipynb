```python
import os
import pandas as pd

lista_arquivo = os.listdir('/content/drive/MyDrive/PYTHON_SENAI/01-20 - Curso de Python - Aula 2/Vendas')

#Criar tabela vazia /
tabela_total = pd.DataFrame()

#importar a base de dados de vendas
for arquivo in lista_arquivo:
  #importar arquivos com pandas (pd)
    if "Vendas" in arquivo:
     tabela = pd.read_csv(f'/content/drive/MyDrive/PYTHON_SENAI/01-20 - Curso de Python - Aula 2/Vendas/{arquivo}')
     tabela_total = pd.concat([tabela_total, tabela])



# Tratar / compilar as bases de dados
display(tabela_total)
```

```python
#Calcular o produto mais vendido
tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[ ['Quantidade Vendida'] ].sort_values(by = 'Quantidade Vendida', ascending= False)  #para aparecere somente a quantidade vendida. # sort_values oordenar valores. # Ascending = Ordem Crescente ou Ordem Decrescente
display(tabela_produtos)
```

```python
#Criar coluna faturamento

tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby('Produto').sum()

tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by = 'Faturamento', ascending= False)
display(tabela_faturamento)
```

```python
#Calcular a loja que mais vendeu em faturamento

tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by = 'Faturamento', ascending= False)
display(tabela_lojas)
```

```python
import plotly.express as px

#importa o plotly para criar graficos, entao cria var grafico e .bar (barras)
grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()
```
